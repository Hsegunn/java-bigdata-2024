- 지도 마커표시
    - Marker() 클래스로 생성한 뒤 지도에 추가
    - icon()내 들어가는 아이콘 속성값은 https://getbootstrap.com/docs/3.3/components/ 에서 확인

````python
# 마커
map = folium.Map(location=[35.178508, 129.199443], tiles='CartoDB Positron', zoom_start=16)
folium.Marker(location=[35.178508, 129.199443], tooltip='송정 해수욕장', popup='서핑하러 오세요').add_to(map)
# popup = folium.Popup('<pre><a href="https://www.starbucks.co.kr/">스타벅스</a>로 오세요</pre>', min_width=200, max_width=300)
popup = folium.Popup('<iframe width="560" height="315" src="https://www.youtube.com/embed/EJF919p_hb0?si=B3lR8OFxMOhA2__I" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay;"></iframe>')
folium.Marker(location=[35.180762, 129.202324], tooltip='스타벅스 부산송정비치점', popup=popup, icon=folium.Icon(color='green', icon='heart')).add_to(map)

map
````
