## 2번 문제

```python
# 2. 일주일의 영어명은 'MON', 'TUE', 'WED', 'TUR', 'FRI', 'SAT', 'SUN' 입니다. 
# 이를 입력하면, '오늘은 월요일~일요일 입니다'라는 메시지가 출력되는 프로그램을 작성하세요.

num = input("요일 입력: ")
if num == 'MON' or num == 'mon':
    print('오늘은 월요일입니다')
elif num == 'TUE' or num == 'tue':
    print('오늘은 화요일입니다')
elif num == 'WED' or num == 'wed':
    print('오늘은 수요일입니다')
elif num == 'THR' or num == 'thr':
    print('오늘은 목요일입니다')
elif num == 'FRI' or num == 'fri':
    print('오늘은 금요일입니다')
elif num == 'SAT' or num == 'sat':
    print('오늘은 토요일입니다')
elif num == 'SUN' or num == 'sun':
    print('오늘은 일요일입니다')  
else:
    print('*오류발생, 영어로 MON, TUE, WED, TUR, FRI, SAT, SUN 로 입력해야 합니다')
```

## 3번 문제

```python
# 3. 나이를 입력하면 19세 미만일 경우 입자을 못하게 하는 클럽 입장 프로그램을 작성하세요.
# 예) 나이 입력
# 17 -> '애들은 가라'
# 24 -> '어서옵쇼, 손님!!'

age = int(input('나이: '))

if age < 19:
    print('애들은 가라')
elif age >= 19:
    print('어서옵쇼, 손님!!')
else:
    print('잘못된 값이 입력되었습니다.')
```

## 4번 문제

```python
# 4. while 문을 사용, 1부터 1000 까지 에서 3과 5의 배수를 모두 구한뒤,
# result.txt에 저장하는 프로그램을 작성하세요.

res = 1
# 주피터 노트북은 파일명만 써도 현재위치에 파일이 생성됨, 파이썬은'./폴더위치/result.txt'
with open('result.txt', mode='w', encoding='utf-8') as f:
    while res <= 1000:
        if res % 3 ==0:     
            print(f'3의배수:{res}', end=' ')
            f.write(f'3의배수:{res}\n')
        elif res % 5 ==0:
            print(f'5의배수:{res}', end=' ')
            f.write(f'5의배수:{res}\n')
        res += 1
```
